```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# Add the image name on the variable name face 
# It will add the image name to all the face variables
# %store will store the face variables and let it share the link to all the cells 
print("What face do you want to try [1/8]")
N = input(" > ")
pic = "Face(" + N + ").jpg"

face = "Faces/" + pic
%store face

# Display the chosen picture 
img = cv2.imread(face)
plt.imshow(img[:, :, ::-1])
```

```python
# Analyze the chosen face
img = face
demography = DeepFace.analyze(img, ['age', 'gender', 'race', 'emotion'])

# Check if the user what the Statistical analysis of the face
print("Do you want to to see your demography")
yn = input("[Y/N]")

if yn == "y":
    print(demography)
```

```python
# Display the Facial Attribute of the face
img_path = face
img = cv2.imread(img_path)
plt.imshow(img[:, :, ::-1])

print("Age:", demography[0]["age"])
print("Gender:", demography[0]["dominant_gender"])
print("Race:", demography[0]["dominant_race"])
print("Emotion:", demography[0]["dominant_emotion"])
```

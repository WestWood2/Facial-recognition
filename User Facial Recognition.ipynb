```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt


key = cv2. waitKey(1)
webcam = cv2.VideoCapture(0)
print("< This is a Facial recognition program >\n=====================================================\n> to start click on the screen with your face \n> Then press 'S' to take a Picture or 'Q' to End the program\n> Then let the code to analyze your face")
print("\n< !!Please note!! >\n=====================================================\n> make sure user face is visible\n> make sure the camera clean and clear \n> face need to be close to the camera to get best results.")
while True:
    try:
        check, frame = webcam.read()
        cv2.imshow("Capturing", frame)
        key = cv2.waitKey(1)
        if key == ord('s'): 
            cv2.imwrite(filename='face_img.jpg', img=frame)
            webcam.release()
            cv2.waitKey(1650)
            cv2.destroyAllWindows()
            face = "face_img.jpg"
            img = cv2.imread(face)
            plt.imshow(img[:, :, ::-1])
            img = "face_img.jpg"
            print("")
            demography = DeepFace.analyze(img, ['age', 'gender', 'race', 'emotion'])
            print("Age:", demography[0]["age"])
            print("Gender:", demography[0]["dominant_gender"])
            print("Race:", demography[0]["dominant_race"])
            print("Emotion:", demography[0]["dominant_emotion"])
            break
        elif key == ord('q'):
            webcam.release()
            cv2.destroyAllWindows()
            break

           
    except(KeyboardInterrupt):
        print("Turning off camera.")
        webcam.release()
        print("Camera off.")
        print("Program ended.")
        cv2.destroyAllWindows()
        break
    
```
